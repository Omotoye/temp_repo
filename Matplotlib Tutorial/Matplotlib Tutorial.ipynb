# Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

## Resizing the Graph
plt.figure(figsize=(5,3), dpi=150)

plt.plot(x, y, label='2x', color='b', linewidth=2, marker='o', markersize=5, markeredgecolor='red', linestyle='--')

fontdict = {
    'fontname': 'Impact',
    'fontsize': 20
}
label_font = {
    'fontname': 'Franklin Gothic Medium', 
    'fontsize': 15
}

## Line Number Two
x2 = np.arange(0,4.5,0.5)
plt.plot(x2, x2**2, 'yx-', markeredgecolor='b', label='x^2')

plt.title('Our First Graph', fontdict=fontdict)
plt.xlabel('X Axis', fontdict=label_font)
plt.ylabel('Y Axis', fontdict=label_font)

plt.xticks(range(0,4))
plt.yticks(range(0,12,2))

plt.legend()
# Using shorthand notation
# fmt = '[color][marker][line]'
plt.figure(figsize=(6,2),dpi=160)
plt.subplot(121)
plt.plot(x,y,'ro-', label='2y')
plt.legend()
plt.subplot(122)
plt.plot(x2, x2**2, 'yx-', markeredgecolor='b', label='x^2')
# Adding a Legend
plt.legend()

plt.show()
```

## Bar Chart

```python
plt.figure(figsize=(6,4), dpi=150)

labels = ['A', 'B', 'C']
values = [1, 4, 2]

bars = plt.bar(labels, values)
bars[0].set_hatch('/')
bars[0].set_label('First')
bars[1].set_hatch('o')
bars[1].set_label('Second')
bars[2].set_hatch('*')
bars[2].set_label('Last')
plt.show()
```

## Line Graph

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(dpi=129)
plt.title('Gas Prices over Time (in USD)', fontdict=fontdict)

plt.plot(gas.Year, gas.USA, label='USA')
plt.plot(gas.Year, gas.Canada, label='Canada')
plt.plot(gas.Year, gas['South Korea'], label='South Korea')
plt.fill_between(gas.Year, gas['South Korea'], gas.Canada)
plt.fill_between(gas.Year, gas.Canada, gas.USA, alpha=0.6)
plt.fill_between(gas.Year, gas.USA)

plt.xticks(gas.Year[::3])
plt.xlabel('Year', fontdict=label_font)
plt.ylabel('US Dollars', fontdict=label_font)
plt.legend()

plt.figure(dpi= 150)
plt.title('Gas Prices over Time (in USD)', fontdict=fontdict)
for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], label=country)
plt.xticks(gas.Year[::3])
plt.xlabel('Year', fontdict=label_font)
plt.ylabel('US Dollars', fontdict=label_font)
plt.legend(loc=2, prop={'size':5.5})
plt.savefig('Gas_price_figure.png', dpi=300)
plt.show()

```

## LivePlots

```python
import random
from matplotlib.animation import FuncAnimation
import matplotlib.pyplot as plt
from itertools import count
%matplotlib inline

plt.style.use('fivethirtyeight')
x_vals = []
y_vals = []

index = count()

def animate(i):
    x_vals.append(next(index))
    y_vals.append(random.randint(0,5))
    plt.plot(x_vals, y_vals)
    
ani = FuncAnimation(plt.gcf() , animate, interval=1000)
    
plt.tight_layout()
plt.show()
```

```python
from matplotlib.animation import FuncAnimation
from itertools import count
import random
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
x_vals = []
y_vals = []

index = count()

def animate(i):
    x_vals.append(next(index))
    y_vals.append(random.randint(0,5))
    plt.cla()
    plt.plot(x_vals, y_vals)
    
    
ani = FuncAnimation(plt.gcf() , animate, interval=1000)
    
plt.tight_layout()
plt.show()
```
