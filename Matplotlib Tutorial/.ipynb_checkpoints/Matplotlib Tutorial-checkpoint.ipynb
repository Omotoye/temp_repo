# Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

## Resizing the Graph
plt.figure(figsize=(5,3), dpi=150)

plt.plot(x, y, label='2x', color='b', linewidth=2, marker='o', markersize=5, markeredgecolor='red', linestyle='--')

fontdict = {
    'fontname': 'Impact',
    'fontsize': 20
}
label_font = {
    'fontname': 'Franklin Gothic Medium', 
    'fontsize': 15
}

## Line Number Two
x2 = np.arange(0,4.5,0.5)
plt.plot(x2, x2**2, 'yx-', markeredgecolor='b', label='x^2')

plt.title('Our First Graph', fontdict=fontdict)
plt.xlabel('X Axis', fontdict=label_font)
plt.ylabel('Y Axis', fontdict=label_font)

plt.xticks(range(0,4))
plt.yticks(range(0,12,2))

plt.legend()
# Using shorthand notation
# fmt = '[color][marker][line]'
plt.figure()
plt.plot(x,y,'ro-', label='2y')

# Adding a Legend
plt.legend()

plt.show()
```

### Bar Chart

```python
plt.figure(figsize=(6,4))

labels = ['A', 'B', 'C']
values = [1, 4, 2]

bars = plt.bar(labels, values)
bars[0].set_hatch('/')
bars[0].label='First'
bars[1].set_hatch('o')
bars[2].set_hatch('*')
plt.legend()
plt.show()
```
