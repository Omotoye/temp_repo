```python
import matplotlib.pyplot as plt
from matplotlib.animation import  FuncAnimation
from matplotlib.lines import Line2D
from IPython import display
import numpy as np
%matplotlib widget

x = np.linspace(0, 2*np.pi, 100)

# The first thing to do when creating an Animation is 
# to create an object of the figure that the plot would 
# be animated upon
fig = plt.figure()

# Next you create an empty plot and take the line object
# so the animation can manipulate it 
# lines = plt.plot([])
# line = lines[0]

ax = fig.add_axes([0.1,0,0.9,0.9])
#line, = ax.plot([])
line = Line2D([],[])
ax.add_line(line)

## Another way is to use tuple unpacking 
# line, = plt.plot([])

# Here you call other setup of what you want from you plot
# like the x and y limit, some static line that's not supposed
# to change.
ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-1.1, 1.1)

# Next make  a function that's going to be called by the funcAnimate
# function. 

def animate(frame):
    # update plot
    y = np.sin(x + 2*np.pi * frame/100)
    line.set_data((x,y))

anim = FuncAnimation(fig, animate, frames=100, interval=20)
video = anim.to_html5_video()
html = display.HTML(video)
display.display(html)
plt.close()
```
